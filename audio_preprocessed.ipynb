```python
import librosa 
import warnings
warnings.filterwarnings('ignore')
import soundfile as sf
import os
import moviepy.editor as mp
import math
```

```python
cnt = 0
segment_duration = 300
for files in os.listdir("/content/who-s-talking/audios")[6:]:
  os.mkdir(f'/content/drive/MyDrive/Colab Notebooks/audio/{files}')
  for aud in os.listdir(f'/content/who-s-talking/audios/{files}'):
    audio = mp.AudioFileClip(f'/content/who-s-talking/audios/{files}/{aud}')
    audio_duration = audio.duration
    num_segments = math.ceil(audio_duration / segment_duration)

    for i in range(num_segments):
      start_time = i * segment_duration
      end_time = min((i + 1) * segment_duration, audio_duration)
      segment = audio.subclip(start_time, end_time)
      segment.write_audiofile(f'/content/drive/MyDrive/Colab Notebooks/audio/{files}/{cnt}{files}{i}.wav')
      cnt += 1
      print(cnt)
```
